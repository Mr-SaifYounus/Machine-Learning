#importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

#Step 1:Generating synthetic data

```python
#generate synthetic data for example
np.random.seed(0)
hours_studied=np.random.uniform(0,10,100)
exam_scores=50+5*hours_studied+np.random.normal(0,2,100)
print("Studied hours :",hours_studied,)
print("="*50)
print("The type of studied hours",type(hours_studied))
print("="*50)
print("Exam Scores are :",exam_scores)
print("="*50)
print("Type of Exam scores",type(exam_scores))
```

#Step2: Converting synthetic data in dataframe for columns

```python
import pandas as pd
data=pd.DataFrame({"Hours Studied":hours_studied,"Exam Scores":exam_scores})
```

```python
data.head(100)
```

#Step 3:Splitting data for training and testing

```python
#split the data into training and testing (80%training,20%testing)
x=data[["Hours Studied"]]
y=data[["Exam Scores"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_test)
print("="*50)
print("Length of x test is",len(x_test))
print(y_test,len(y_test))
```

```python
print(y_test,"\n","\n",len(y_test))
print(x_test.ndim)
```

```python
print(x_train)
print(y_train)
```

#Step 4: Create ansd train a linear regression model

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

#Step 5: MODEL EVALUATION

```python
#MAKE PREDICTIONS ON TESTING DATA
y_pred=model.predict(x_test)
```

```python
#Calculate mean squared error (MSE) between the two variables and predicted values
mse=mean_squared_error(y_test,y_pred)
#calculate the (R^2) rsquare scored between true values and predicted values
r2=r2_score(y_test,y_pred)
#computting mean squared error with formatting
print(f"Mean squared error:{mse:.2f}")
#print(r squared value with formatting)
print(f"R squared: {mse:.2f}")
```

```python
x_test = x_test.to_numpy().reshape(-1)
y_test = y_test.to_numpy().reshape(-1)
y_pred = y_pred.to_numpy().reshape(-1)
```
